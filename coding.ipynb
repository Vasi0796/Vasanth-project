# Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder

# Create a label encoder object
label_encoder = LabelEncoder()

# List of categorical columns to encode
categorical_columns = ['EmploymentStatus', 'EducationLevel', 'MaritalStatus', 'HomeOwnershipStatus', 'LoanPurpose']

# Apply label encoder to each categorical column
for col in categorical_columns:
    data[col] = label_encoder.fit_transform(data[col])
```

# Data Splitting for Classification Models 

```python
from sklearn.model_selection import train_test_split

# Define the target variable (LoanApproved)
target = 'LoanApproved'

# Define the features based on correlation with LoanApproved
features = [
    'MonthlyIncome',
    'AnnualIncome',
    'NetWorth',
    'TotalAssets',
    'CreditScore',
    'Age',
    'Experience',
    'LengthOfCreditHistory'
]
```

```python
# Split the data into features (X) and target (y)
X = data[features]  
y = data[target]    
```

```python
# Checking the class distribution
class_counts = y.value_counts()

# Print class distribution
print("Class distribution:")
print(class_counts)

# Calculate percentage distribution
class_percentage = (class_counts / len(y)) * 100
print("\nPercentage distribution:")
print(class_percentage)

# Visualizing the class distribution
plt.figure(figsize=(8, 5))
sns.barplot(x=class_counts.index, y=class_counts.values, palette="viridis")
plt.title("Class Distribution in Target Variable", fontsize=16)
plt.xlabel("Class", fontsize=14)
plt.ylabel("Count", fontsize=14)
for i, count in enumerate(class_counts.values):
    plt.text(i, count + 0.02 * max(class_counts.values), str(count), ha='center', fontsize=12)
plt.show()
```

```python
# Using SMOTE for making the class balance
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Applying SMOTE to the training set
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Checking class distribution after applying SMOTE
print("Class distribution before SMOTE:")
print(y_train.value_counts())

print("\nClass distribution after SMOTE:")
print(y_train_resampled.value_counts())
```

```python
# Checking the class balance
# Visualizing the class distribution after SMOTE
plt.figure(figsize=(8, 5))
sns.barplot(x=y_train_resampled.value_counts().index, 
            y=y_train_resampled.value_counts().values, 
            palette="viridis")
plt.title("Class Distribution After SMOTE", fontsize=16)
plt.xlabel("Class", fontsize=14)
plt.ylabel("Count", fontsize=14)
for i, count in enumerate(y_train_resampled.value_counts().values):
    plt.text(i, count + 0.02 * max(y_train_resampled.value_counts().values), 
             str(count), ha='center', fontsize=12)
plt.show()
```

```python
# Perform the data split (70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Verify the shapes of the resulting splits
print(f"Training Features Shape: {X_train.shape}")
print(f"Testing Features Shape: {X_test.shape}")
print(f"Training Target Shape: {y_train.shape}")
print(f"Testing Target Shape: {y_test.shape}")
```

# Classification Models Evaluation

## 1. Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Initialize the DecisionTreeClassifier
dt_classifier = DecisionTreeClassifier(random_state=42)
```

```python
# Train the model on the training data
dt_classifier.fit(X_train, y_train)
```

```python
# Predict the target variable on the test data
y_pred_dt = dt_classifier.predict(X_test)
```

```python
# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred_dt)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
# Generate the classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred_dt))
```

```python
# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred_dt)

# Plot the Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='rocket', xticklabels=['Not Approved', 'Approved'], yticklabels=['Not Approved', 'Approved'])
plt.title("Confusion Matrix (Decision Tree Classifier)", fontsize=16)
plt.xlabel("Predicted Labels", fontsize=14)
plt.ylabel("True Labels", fontsize=14)
plt.tight_layout()
plt.show()
```

## 3. Stacking Classifier

```python
from sklearn.ensemble import StackingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import ExtraTreesClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression

# Define base models (base learners)
base_learners = [
    ('decision_tree', DecisionTreeClassifier(random_state=42)),
    ('extra_trees', ExtraTreesClassifier(random_state=42)),
    ('random_forest', RandomForestClassifier(random_state=42))
]

# Define meta model (final estimator)
meta_model = LogisticRegression(random_state=42)

# Initialize the Stacking Classifier with the base learners and meta model
stacking_classifier = StackingClassifier(estimators=base_learners, final_estimator=meta_model)
```

```python
# Train the Stacking Classifier on the training data
stacking_classifier.fit(X_train, y_train)
```

```python
# Predict the target variable on the test data
y_pred_stack = stacking_classifier.predict(X_test)
```

```python
# Calculate the accuracy of the model
accuracy_stack = accuracy_score(y_test, y_pred_stack)
print(f"Accuracy: {accuracy_stack * 100:.2f}%")
```

```python
# Generate the classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred_stack))
```

```python
# Confusion Matrix
conf_matrix_stack = confusion_matrix(y_test, y_pred_stack)

# Plot the Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_stack, annot=True, fmt='d', cmap='rocket', xticklabels=['Not Approved', 'Approved'], yticklabels=['Not Approved', 'Approved'])
plt.title("Confusion Matrix (Stacking Classifier)", fontsize=16)
plt.xlabel("Predicted Labels", fontsize=14)
plt.ylabel("True Labels", fontsize=14)
plt.tight_layout()
plt.show()
```

# Model Comparison for loan approval forecasting

```python
# Accuracy results from each model
models = ['Decision Tree', 'Extra Trees', 'Stacking Classifier']
accuracies = [accuracy * 100, accuracy_et * 100, accuracy_stack * 100]

# Create a DataFrame for visualization
model_comparison = {
    'Model': models,
    'Accuracy': accuracies
}

# Plotting the bar chart for model comparison
plt.figure(figsize=(8, 6))
bar_plot = sns.barplot(x=model_comparison['Model'], y=model_comparison['Accuracy'], palette='viridis')

# Add the accuracy values on top of the bars
for i, v in enumerate(accuracies):
    bar_plot.text(i, v + 1, f"{v:.2f}%", ha='center', fontsize=12, color='black')

plt.title("Model Comparison: Accuracy (%)", fontsize=16)
plt.xlabel("Models", fontsize=14)
plt.ylabel("Accuracy (%)", fontsize=14)
plt.tight_layout()
plt.show()

# Conclusion
best_model = models[accuracies.index(max(accuracies))]  

# Display the conclusion
print(f"\nThe best fitted model for loan approval forecasting is the **{best_model}** with an accuracy of {max(accuracies):.2f}%.")
```

# Data Splitting for Regression Models

```python
# Define the features (X) and target variable (y)
features_reg = [
    'BankruptcyHistory', 'TotalDebtToIncomeRatio', 'DebtToIncomeRatio', 
    'InterestRate', 'PreviousLoanDefaults', 'BaseInterestRate', 
    'LoanAmount', 'EmploymentStatus', 'MonthlyLoanPayment', 
    'CreditCardUtilizationRate'
]

# Target variable: RiskScore
X_reg = data[features_reg]  
y_reg = data['RiskScore']  
```

```python
# Split the data into training and testing sets (80% training, 20% testing)
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

# Display the shape of the splits to confirm
print(f"Training data shape (X_train, y_train): {X_train_reg.shape}, {y_train_reg.shape}")
print(f"Testing data shape (X_test, y_test): {X_test_reg.shape}, {y_test_reg.shape}")
```

## 1. Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Initialize the Random Forest Regressor
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
# Fit the model
rf_regressor.fit(X_train_reg, y_train_reg)
```

```python
# Predict on the test data
y_pred_rf = rf_regressor.predict(X_test_reg)
```

```python
# Calculate R-squared, MSE, RMSE, and MAE
r2_rf = r2_score(y_test_reg, y_pred_rf)
mse_rf = mean_squared_error(y_test_reg, y_pred_rf)
rmse_rf = np.sqrt(mse_rf)
mae_rf = mean_absolute_error(y_test_reg, y_pred_rf)

# Print the evaluation metrics
print(f"R-squared: {r2_rf:.4f}")
print(f"Mean Squared Error (MSE): {mse_rf:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse_rf:.4f}")
print(f"Mean Absolute Error (MAE): {mae_rf:.4f}")
```

```python
# Plot Actual vs Predicted values
plt.figure(figsize=(10,6))
plt.scatter(y_test_reg, y_pred_rf, color='blue', alpha=0.6)
plt.plot([min(y_test_reg), max(y_test_reg)], [min(y_test_reg), max(y_test_reg)], color='red', linestyle='--')
plt.xlabel('Actual RiskScore')
plt.ylabel('Predicted RiskScore')
plt.title('Actual vs Predicted RiskScore (Random Forest Regression)')
plt.grid(True)
plt.show()
```

## 2. Gradient Boosting Regression

```python
from sklearn.ensemble import GradientBoostingRegressor

# Initialize the Gradient Boosting Regressor
gb_regressor = GradientBoostingRegressor(n_estimators=100, random_state=42)
```

```python
# Predict on the test data
y_pred_gb = gb_regressor.predict(X_test_reg)
```

```python
# Calculate R-squared, MSE, RMSE, and MAE
r2_gb = r2_score(y_test_reg, y_pred_gb)
mse_gb = mean_squared_error(y_test_reg, y_pred_gb)
rmse_gb = np.sqrt(mse_gb)
mae_gb = mean_absolute_error(y_test_reg, y_pred_gb)

# Print the evaluation metrics
print(f"R-squared: {r2_gb:.4f}")
print(f"Mean Squared Error (MSE): {mse_gb:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse_gb:.4f}")
print(f"Mean Absolute Error (MAE): {mae_gb:.4f}")
```

```python
# Plot Actual vs Predicted values for Gradient Boosting Regression
plt.figure(figsize=(10,6))
plt.scatter(y_test_reg, y_pred_gb, color='green', alpha=0.6)
plt.plot([min(y_test_reg), max(y_test_reg)], [min(y_test_reg), max(y_test_reg)], color='red', linestyle='--')
plt.xlabel('Actual RiskScore')
plt.ylabel('Predicted RiskScore')
plt.title('Actual vs Predicted RiskScore (Gradient Boosting Regression)')
plt.grid(True)
plt.show()
```

# Model Comparison for Risk Score Forecasting

```python
# MSE, RMSE, MAE, and R-squared values for each model
mse_scores = [mse_rf, mse_gb, mse_xgb]
rmse_scores = [rmse_rf, rmse_gb, rmse_xgb]
mae_scores = [mae_rf, mae_gb, mae_xgb]
r2_scores = [r2_rf, r2_gb, r2_xgb]
models = ['Random Forest', 'Gradient Boosting', 'XGBoost']

# MSE Plot
plt.figure(figsize=(8, 6))
plt.bar(models, mse_scores, color=['blue', 'green', 'purple'], alpha=0.7)
plt.title('Model Comparison: MSE for RiskScore Forecasting')
plt.xlabel('Models')
plt.ylabel('Mean Squared Error (MSE)')
for i, v in enumerate(mse_scores):
    plt.text(i, v + 0.5, f'{v:.4f}', ha='center', va='bottom', fontsize=12)
plt.tight_layout()
plt.grid(True)
plt.show()

# RMSE Plot
plt.figure(figsize=(8, 6))
plt.bar(models, rmse_scores, color=['blue', 'green', 'purple'], alpha=0.7)
plt.title('Model Comparison: RMSE for RiskScore Forecasting')
plt.xlabel('Models')
plt.ylabel('Root Mean Squared Error (RMSE)')
for i, v in enumerate(rmse_scores):
    plt.text(i, v + 0.2, f'{v:.4f}', ha='center', va='bottom', fontsize=12)
plt.tight_layout()
plt.grid(True)
plt.show()

# MAE Plot (Adjusting the label position to prevent overlap)
plt.figure(figsize=(8, 6))
plt.bar(models, mae_scores, color=['blue', 'green', 'purple'], alpha=0.7)
plt.title('Model Comparison: MAE for RiskScore Forecasting')
plt.xlabel('Models')
plt.ylabel('Mean Absolute Error (MAE)')
for i, v in enumerate(mae_scores):
    plt.text(i, v + 0.1, f'{v:.4f}', ha='center', va='bottom', fontsize=12)  # Adjusted label position
plt.tight_layout()
plt.grid(True)
plt.show()

# R-squared Plot (Adding count values to the graph)
plt.figure(figsize=(8, 6))
plt.bar(models, r2_scores, color=['blue', 'green', 'purple'], alpha=0.7)
plt.title('Model Comparison: R-squared for RiskScore Forecasting')
plt.xlabel('Models')
plt.ylabel('R-squared')
for i, v in enumerate(r2_scores):
    plt.text(i, v + 0.01, f'{v:.4f}', ha='center', va='bottom', fontsize=12)  # Added count labels
plt.tight_layout()
plt.grid(True)
plt.show()

# Conclusion based on the MSE, RMSE, MAE, and R-squared values
print(f"\nComparison of Models based on MSE, RMSE, MAE, and R-squared:")
print(f"Random Forest - MSE: {mse_rf:.4f}, RMSE: {rmse_rf:.4f}, MAE: {mae_rf:.4f}, R-squared: {r2_rf:.4f}")
print(f"Gradient Boosting - MSE: {mse_gb:.4f}, RMSE: {rmse_gb:.4f}, MAE: {mae_gb:.4f}, R-squared: {r2_gb:.4f}")
print(f"XGBoost - MSE: {mse_xgb:.4f}, RMSE: {rmse_xgb:.4f}, MAE: {mae_xgb:.4f}, R-squared: {r2_xgb:.4f}")

# Determine the best model based on MSE, RMSE, MAE, and R-squared (lower MSE, RMSE, MAE; higher R-squared is better)
best_model = min([(r2_rf, mse_rf, rmse_rf, mae_rf, "Random Forest"),
                  (r2_gb, mse_gb, rmse_gb, mae_gb, "Gradient Boosting"),
                  (r2_xgb, mse_xgb, rmse_xgb, mae_xgb, "XGBoost")], key=lambda x: (x[1], x[2], x[3], -x[0]))

print(f"\nConclusion based on evaluation metrics: The best model is {best_model[4]} with the following values:")
print(f"MSE: {best_model[1]:.4f}, RMSE: {best_model[2]:.4f}, MAE: {best_model[3]:.4f}, R-squared: {best_model[0]:.4f}")
```
